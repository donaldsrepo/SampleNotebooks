```python
import pandas as pd
df = pd.read_csv('sample_data_small.txt')
```

```python
df
```

```python
df[df['Name'].str.contains('\D')]
```

```python
df.groupby(['Sex','Job']).agg({'Age': ['sum'], 'Name': pd.Series.mode})
```

```python
df2 = df.groupby(['Sex'], as_index=False).agg({'Age': 'sum', 'Name': 'min', 'Job':  pd.Series.mode})
df2

#df.groupby('Column1').agg({'Column3': ['sum'], 'Column4': ['sum'], 'Column5': ['sum'], 'Column6': pd.Series.mode, 'Column7': pd.Series.mode, 'Column8': pd.Series.mode}) 
```

```python
df2 = df.groupby(['Sex'], as_index=False).agg({'Age': 'sum', 'Name': pd.Series.mode, 'Job':  pd.Series.mode})
df2
```

```python
df2.columns = df2.columns.to_flat_index()
df2
```

```python
df2.columns = list(map(''.join, df2.columns.values))
df2
```

```python
df.groupby(['Sex','Job'], as_index=False).agg({'Age': ['sum'], 'Name': 'min'})
```

```python
df2 = df.groupby(['Sex','Job']).agg({'Age': ['sum'], 'Name': 'min'})
```

```python
df2.columns
```

```python
df2.columns = list(map(''.join, df2.columns.values))
```

```python
df2.to_flat_index()
```

```python
df.groupby(['Sex']).agg({'Age': 'sum'})
```

```python
df.groupby(['Sex','Job']).agg({'Age': 'sum', 'Name': 'min'})
```

```python
df.groupby('Team').agg({'Age': ['sum'], 'Pos': ['sum']})
```

```python
df.groupby(['Team']).agg({'Age','Pos': 'sum'})
```

```python
(df.groupby(['orig','dest','dep_date','ret_date','airline'], as_index = False, sort=False)
   .agg({'search_date': 'first', 'price': 'min'})
   .reset_index(drop = True))
```
