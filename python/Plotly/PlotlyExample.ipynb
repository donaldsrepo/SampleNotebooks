```python
# Standard plotly imports
import chart_studio.plotly as py
import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode
# Using plotly + cufflinks in offline mode
import cufflinks
cufflinks.go_offline(connected=True)
init_notebook_mode(connected=True)
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
import os
import sys
```

```python
boston = load_boston()
df = pd.DataFrame(boston.data, columns=boston.feature_names)
y = boston.target
#df=pd.read_csv('C:\\DataScience\\boston_house_prices.csv', sep=',',header=[0])
```

```python
import plotly.io as pio
pio.renderers
```

```python
# test environment to see if running in colab or not
colab = os.environ.get('COLAB_GPU', '10')
if (int(colab) == 0):
  renderer_type = "colab"
  print('colab found using render type ', renderer_type)
else:
  renderer_type = "plotly_mimetype+notebook_connected"
  print('colab not found using render type ', renderer_type)
```

```python
df['RAD_CAT']=df['RAD'].astype(str)
df
```

```python
fig = go.Figure(data=[go.Histogram(x=df['TAX'] )])

fig.update_layout(
    title_text='TAX Distribution', # title of plot
    xaxis_title_text='TAX', # xaxis label
    yaxis_title_text='Count' # yaxis label
)

fig.show(renderer=renderer_type)
```

```python
df['TAX'].iplot(kind='hist', xTitle='TAX',
                  yTitle='count', title='TAX Distribution')
```

```python
df[['CRIM', 'INDUS']].iplot(
    kind='hist',
    histnorm='percent',
    barmode='overlay',
    xTitle='Value',
    yTitle='(%) of Data',
    title='CRIM and INDUS')
```

```python
df.pivot(columns='RAD', values='CRIM').iplot(
        kind='box',
        yTitle='CRIM',
        title='CRIM by RAD')
```

```python
df.iplot(
    x='CRIM',
    y='INDUS',
    # Specify the category
    categories='RAD_CAT',
    xTitle='CRIM',
    yTitle='INDUS',
    title='CRIM vs INDUS by RAD')
```

```python
df.iplot(
    x='CRIM',
    y='INDUS',
    #size=df['RAD'].to_numpy(),
    size=df['RAD']/1,
    #size=1,
    text='RAD',
    mode='markers',
    # Log xaxis
    layout=dict(
        xaxis=dict(type='log', title='CRIM'),
        yaxis=dict(title='INDUS'),
        title='CRIM vs INDUS Sized by RAD'))
```

```python
import plotly.figure_factory as ff
figure = ff.create_scatterplotmatrix(
    df[['CRIM', 'INDUS', 'RAD_CAT',      
        'AGE','TAX']],
    diag='histogram',
    index='RAD_CAT')
```

```python
display(figure)
```

```python
corrs = df.corr()
figure2 = ff.create_annotated_heatmap(
    z=corrs.values,
    x=list(corrs.columns),
    y=list(corrs.index),
    annotation_text=corrs.round(2).values,
    showscale=True)
```
