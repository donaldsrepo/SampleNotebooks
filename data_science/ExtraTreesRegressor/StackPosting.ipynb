```python
import pandas as p
import numpy as np
from sklearn.model_selection import train_test_split

classes = {0: 'Other', 1: 'Thunder'}

x_new = p.read_csv('train.csv', sep=',').values # x_new: <class 'numpy.ndarray'>
y_predict = train_test_split(x_new) # y_predict: <class 'list'>

z = np.asarray(y_predict)[0].T
for i in range(0,len(z)):
    print(z[12][i])
    print(classes[z[12][i]])
    
#z = y_arr[0].T
print(z[12])


print(classes[z[12][0]]) # y_predict[0] <class 'numpy.ndarray'> > [[5254 106.47 'ay' ... 0 0 0] [4040 107.61 'ay' ... 0 0 0]
print(classes[y_predict[1]])
```
